```python
## 1. Problem Statement

The goal of this task is to explore and visualize the Iris dataset in order to
understand its structure, feature distributions, and relationships between
different variables using data analysis and visualization techniques.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.shape
```

```python
iris.columns
```

```python
iris.info()
```

```python
iris.describe()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='petal_length',
    hue='species'
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
iris.hist(figsize=(10,6))
plt.suptitle("Feature Distributions")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Features")
plt.show()
```

```python
## 7. Final Insights

- The Iris dataset contains 150 samples and 5 columns with no missing values.
- Petal features show clearer separation between flower species compared to sepal features.
- Some features contain mild outliers, which can be observed in box plots.
- Visualization helps in understanding patterns and relationships before model building.
```
